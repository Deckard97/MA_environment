```python
import pandas as pd
```

```python
df = pd.read_csv('dataset.csv')
```

```python
column_names = df.columns
column_names_list = df.columns.tolist()
```

```python
column_names_list
```

```python
# Create an empty dictionary to store the new column names
rename_dict = {}

# Iterate through each column and ask for a new name
for column in df.columns:
    new_name = input(f"Enter new name for column '{column}' (press enter to keep original): ")
    if new_name:
        rename_dict[column] = new_name
    else:
        rename_dict[column] = column

# Rename the columns using the dictionary
df.rename(columns=rename_dict, inplace=True)

# Print the DataFrame with new column names (optional)
print(df)
```
